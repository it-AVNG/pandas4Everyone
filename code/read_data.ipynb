```python
url ='https://jsonplaceholder.typicode.com/users'
users = pd.read_json(
    url,
    
)
```

```python
users
```

```python
import json
from pandas import json_normalize
users_add = json_normalize(users['address'])
users_add= users_add.add_prefix('address_')
```

```python
users_company =json_normalize(users.company)
users_company=users_company.add_prefix('company_')
```

```python
users= (users
        .join(users_company)
        .join(users_add)
        .drop(cols=['address','company']))
```
