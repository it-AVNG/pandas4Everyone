```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
print(tips)
```

# Univariate
Just like with `matplotlib` we will make a series of univariate plots
## Histogram

```python
hist, ax = plt.subplots()
#add a rug plot for fancy people
sns.rugplot(
    data=tips,
    x='total_bill',
    ax=ax    
)

sns.histplot(
    data=tips,
    x='total_bill',
    ax=ax
    )

ax.set_title('total bill histogram')

plt.show()
```

So far it is a 1-to-1 transition from `matplotlib`. This is due to seaborn is written on top of `matplotlib`, thus creating seamless transition.
Another way to visualize a univariate distribution is using density plot(Kernel Density estimation)
## Density plot
In essence, density plot created by drawing a normal distribution centered at each data point, then smoothing out the overlapping plots so that the area under the curve is 1.

```python
den, ax = plt.subplots()

sns.kdeplot(
    data=tips,
    x='total_bill',
    ax=ax
)

ax.set_title('total bill density')
ax.set_xlabel('total_bill')
ax.set_ylabel('unit probability')
```

for Seaborn, we have a now 'compound' plot at combine histogram and density plot, called Distribution Plots. 
### Distribution Plots
This using `sns.distplot()`which has 2 additional parameter: 
+ `kde`: which is our kernel Density plot
+ `rug`: which is the rug plot

the method returns a data type called 'FacetGrid'

```python
fig = sns.displot(data=tips, x="total_bill", kde=True, rug=True)

fig.set_axis_labels(x_var="Total Bill", y_var="Count")
fig.figure.suptitle('Distribution of Total Bill')

plt.show()
```

## Bar plot(count plot)
bar plot are very similar to histtogram, but instead binning values to produce distribution, bar plot can be used to count discrete variables.
In Seaborn, it is called using `.countplot()`

```python
count, ax = plt.subplots()

# we can use virdis palette to help better distinguish
sns.countplot(data=tips, x='day', palette='viridis', ax=ax)

ax.set_title('Count of Days')
ax.set_xlabel('Day of the week')
ax.set_ylabel('Frequency')
```

# Bivariate Data
We will now use the `seaborn` to plot 2 variable
## Scatter plot
In seaborn, we have 2 type of objects that return a scatter plot, `Axes` and `FacetGrid`
in case that we called `sns.scatterplot()` the methods returns ans `Axes` object

```python
scatting, ax = plt.subplots()

sns.scatterplot(data=tips,x='total_bill',y='tip', ax=ax)

ax.set_title('Scatter Plot of Total Bill and Tip')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Tip')

plt.show()
```

Now, as we have already observed our data, sometimes we would like to see our prediction of the data or to find if there is any relation models between the 2 variable. 
This called for regression line. In seaborn, we called the methods `.regplot()` for this

```python
reg_plot, ax = plt.subplots()
# use fit_reg = False if we just want the scatter
sns.regplot(data=tips, x= 'total_bill', y='tip', ax= ax)

ax.set_title('scatter Plot with regression line of Total Bill and Tip')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Tip')

plt.show()
```

## Joint Plot:
seaborn provide a method to show joining plot that include a univariate plot on each axis: `.jointplot()`
This method returns a `JointGrid` object, so we don't use the `ax` to create a figure. Instead we directly call the underlying `matplotlib.figure()` method.

```python
# joint plot creates the figure and axes for us
joint = sns.jointplot(data=tips, x='total_bill', y= 'tip')

joint.set_axis_labels(xlabel='total bill', ylabel='tip')

# (optional) add suptitle for quality of life
joint.figure.suptitle('Joint plot with total Bill and Tip', y= 1.05)
plt.show()
```

## Hexbin plot
The joint plot above tells us alot of information, but like any story telling, too much details at a wrong time could distract our audiences.
Hexbin plot shows a derived meaningful information without the distraction.
`hexbin()` is a `matplotlib` function, here we use the parameter 'hex' to the joint plot, thus create a hexbin joint plot

```python
hexbin = sns.jointplot(
    data=tips, 
    x='total_bill',
    y='tip',
    kind='hex')

plt.show()
```

## Densityplot of bivariate 
As we know, we can called density plot in univariate field, but what if we pass another parameter for the y axis? - A bivariate density plot\

```python
kd_plot, ax = plt.subplots()
sns.kdeplot(data=tips, x='total_bill',y='tip',ax=ax)

ax.set_title('Kernel Density Plot of Total Bill and Tip')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Tip')

plt.show()
```

joint plot also can provide a kde figure

```python
kde2d = sns.jointplot(data=tips,
                      x='total_bill',
                      y='tip',
                      kind='kde'
                      )
plt.show()
```

## Bar plot
Bar plots can also be used to show multiple variables. By default, `sns.barplot()` Will calculate a mean.

```python
import numpy as np

bar, ax = plt.subplots()

# plot the average total bill for each value of time
# mean is calculated using numpy
sns.barplot(
  data=tips, x="time", y="total_bill", estimator=np.mean, ax=ax
)

ax.set_title('Bar Plot of Average Total Bill for Time of Day')
ax.set_xlabel('Time of Day')
ax.set_ylabel('Average Total Bill')

plt.show()
```

The 'black line' show 95% confidentlevel (where we expecting 95% of values will be located).
Another popular barplot is called boxplot
## Box plot
Unlike the previous plot, box plot shows multiple statistics: the minimum, first quartile,median,thirdquartile, maximum and if applicable, outliers based on the interquartile range and also outliers if possible.

```python
box_plot, ax = plt.subplots()

#y axis is optional, but x would have to be a numeric variable
sns.boxplot(data = tips, x='time', y ='total_bill',ax=ax)

ax.set_title('Box Plot of Total Bill by the time of Day')
ax.set_xlabel('Time of Day')
ax.set_ylabel('total_bill')

plt.show()
```

## Violin plot
A classical visualization of the data is a Violin plot. Violin can show the same information as a box plot. But plot the boxes as kernel density estimation. 

```python
violin_plot, ax =plt.subplots()

sns.violinplot(data=tips, x='time', y='total_bill', ax=ax)

ax.set_title('Violin pot of total bill and time of day')
ax.set_xlabel('Time of day')
ax.set_ylabel('Total bill')

plt.show()
```

Let show the side by side to see the similarity as well as the diffrences

```python
box_violin, (ax1,ax2)= plt.subplots(1,2)

sns.boxplot(data=tips,x='time', y='total_bill',ax= ax1)
sns.violinplot(data=tips,x='time', y='total_bill',ax= ax2)

#set tiltes
ax1.set_title('Box plot')
ax1.set_xlabel('Time of day')
ax1.set_ylabel('Total bill')

ax2.set_title('violin plot')
ax2.set_xlabel('Time of day')
ax2.set_ylabel('Total bill')

# set figure title
box_violin.suptitle('comparision of Boxplot and violin plot')
# compact layout
box_violin.set_tight_layout(True)
```

Another useful graph, personally, I think should be the first to take a look is the pairwise relation ship graph
## Pairwise Relationship
When we have mostly numeric data, visualizing all of the pairwise relationships can be performed using `sns.pairplot()`

```python
fig = sns.pairplot(data=tips)
fig.figure.suptitle(
    'Pairwise Relationship of the Tips Data', y=1.03
)
plt.show()
```

```python
# create a PairGrid, make the diagonal plots on a different scale
pair_grid = sns.PairGrid(tips, diag_sharey=False)

# set a separate function to plot the upper, bottom, and diagonal
# functions need to return an axes, not a figure

# we can use plt.scatter instead of sns.regplot
pair_grid = pair_grid.map_upper(sns.regplot)
pair_grid = pair_grid.map_lower(sns.kdeplot)
pair_grid = pair_grid.map_diag(sns.histplot)

plt.show()
```

## Taking things further 
We can provide multivariate output plot using colors and hue
### Colors
when we are using `sns.violinplot()`, we can use additional parameter called `split` to assigned colors to it

```python
violin, ax = plt.subplots()
sns.violinplot(
    data=tips,
    x='time',
    y='total_bill',
    # using feature as a hue to distinguish
    hue='smoker',
    split=True,
    palette='mako',
    ax = ax
)
plt.show()
```

```python
scatter, ax =plt.subplots()

sns.scatterplot(
    data=tips,
    x='tip',
    y='total_bill',
    hue='smoker',
    palette='plasma',
    alpha=0.7,
    ax=ax
)

plt.show()
```

```python
fig = sns.pairplot(
    data=tips,
    hue='time',
    palette='cividis',
    height=2,
)

plt.show()
```

### Size and Shape
Whorking with point size can be another means of adding more information to a plot.
Use sparingly since human eyes are not very good at comparing areas.

```python
fig, ax =plt.subplots()

sns.scatterplot(
    data=tips,
    x='total_bill',
    y='tip',
    hue='time',
    size='size',
    palette='viridis',
    ax=ax
)

plt.show()
```

## Facet
Creating Facet is when we want to plot multiple plots over a categorical variable
### Univariable Facet
here we will use Anscombe quartet data for illustration

```python
anscombe = sns.load_dataset('anscombe')
print(anscombe)
```

```python
anscombe_plot = sns.relplot(
  
  data=anscombe,
  x="x",
  y="y",
  kind="scatter",
  col="dataset",
  col_wrap=2,
  height=2,
  aspect=1.6, # aspect ratio of each facet
  
)

anscombe_plot.figure.set_tight_layout(True)

plt.show()
```

### Bivariate Facet
additional Categorical variables can be passed into hue, etc parameter

```python
colors = {
    "Yes":"#F9C26F",
    "No":"#4B386B"
}

# make the faceted scatter plot
# the essential to draw the figure

facet2 = sns.relplot(
    data=tips,
    x='tip',
    y='total_bill',
    hue='smoker',
    style='sex',
    kind='scatter',
    col='day',
    row='time',
    palette=colors,
    height=1.7
)

# below is to make the plot pretty
# adjust facet titles
facet2.set_titles(
  row_template=" {row_name}",
  col_template=" {col_name}"
)

# adjust the legend to not have it overlap the figure
sns.move_legend(
  facet2,
  loc="lower center",
  bbox_to_anchor=(0.5, 1),
  ncol=2,   #number legend columns 
  title=None,   #legend title
  frameon=False, #remove frame (i.e., border box) around legend
)

facet2.figure.set_tight_layout(True)
```
